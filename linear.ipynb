2. Determinants and Inverses
Find the determinant of A.

Determine whether A is invertible. If yes, calculate A−1.

```python
import numpy as np

# Define matrix A
A = np.array([
    [8, 6],
    [3, 2]]
    )

# Finding determinant
determinant_A = np.linalg.det(A)
print("Determinant of A:", determinant_A)

# Checking if A is inversible
if determinant_A != 0:
    A_inv = np.linalg.inv(A)
    print("Inverse of A:\n", A_inv)
else:
    print("Matrix A is not inversivle")
```

4. You are given a dataset of cars with the following features:Number of cylinders (x1​)Engine size in liters (x2​)Horsepower (x3)
car 1:

x1: 4

x2: 1.5

x3: 100

and price: 20

car 2:

x1: 6

x2: 2.0

x3: 150

and price: 30

Then, find the price of a car 3 with features

x1: 8

x2: 3.0

x3: 200

price:?

```python
print(" Feature Matrix:")
X=np.array([
    [4,1.5,100],
    [6,2,150]
])
print(X)
print("Price Vector:")
y=np.array([
    [20],
    [30]
])
print(y)
# calculating the  coefficient using pseudo inverse
beta=np.linalg.pinv(X).dot(y)
#finding the price of car3
print("Feature Matrix of car3")
X1=np.array([
    [8,3,200]
])
print(X1)
priceofcar3= X1 @ beta
print("The price of car 3 is")
print(priceofcar3)
```
