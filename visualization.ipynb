```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("health_heart_experimental.csv")

# Display first 5 rows
print(df.head())
```

## Line Graph

```python
# Group by age and take mean SysBP
df_line = df.groupby('age')['SysBP'].mean().reset_index()

plt.figure()
plt.plot(df_line['age'], df_line['SysBP'], color='green')

plt.title("Average Systolic Blood Pressure vs Age")
plt.xlabel("Age")
plt.ylabel("Average Systolic Blood Pressure")
plt.show()
```

The line graph explains that there is a direct upward trend showing that as age increases, average systolic blood pressure also rises. People from the age range of 60 experiences more rise in their blood perssure than from other age group.The steepness of the line indicates how rapidly blood pressure escalates between the age groups.

## Bar Plot

```python
plt.figure()
df['sex'].value_counts().plot(kind='bar')
plt.xlabel("Sex (0 = Female, 1 = Male)")
plt.ylabel("Count")
plt.title("Sex Distribution")
plt.show()
```

The bar plot reveals the sample distribution by gender denoting 0 as female and 1 as male, showing the total count of male versus female participants in the dataset. There is an equal number of 35000 people among male and female which explains that there is no baissness in blood pressure analysis among any specific gender group.

## Pie Chart

```python
plt.figure()
df['indication'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)
plt.title("Heart Disease Indication Percentage")
plt.ylabel("")
plt.show()
```

The pie chart explains which gender experiences more heart disease. It reveals that 52.7% of male participating in the dataset experiences heart diseases which is more than the number of female experiencing it.

## Box Plot

```python
plt.figure()
plt.boxplot(df['BMI'])
plt.ylabel("BMI")
plt.title("BMI Distribution")
plt.show()
```

The box plot shows that the male population has a BMI ranging from roughly 24 to 35 which is the case of normal to class I obesity.The orange line indicates that the median BMI is approximately 29 which is slightly overweight. 

## Histogram

```python
plt.figure()
plt.hist(df['age'], bins=30, edgecolor='black')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
```

The histogram shows that the dataset has a high frequency of participants across all ages from 20 to 70 . Each bar reaches roughly between 2,500 and 3,000 counts explaining that there is high level of participation between all the age groups. There is a noticable gap in number of participants among certain age groups keeping age group range of 34-36 people least involved.

## Scatter Plot

```python
# Aggregate data
df_scatter = df.groupby('age')['SysBP'].mean().reset_index()

plt.figure()
plt.scatter(
    df_scatter['age'],
    df_scatter['SysBP'],
    s=50
)

plt.title("Average Systolic Blood Pressure vs Age")
plt.xlabel("Age")
plt.ylabel("Average Systolic Blood Pressure")
plt.show()
```

The scatter plot shows that there is a subtle upward trend as the points move from age 20 toward age 70. The  average values fall between 125.2 and 126.0 mmHg. The highest average pressure recorded is for the 61-year-old group, while the lowest average occurs around age 31 .
